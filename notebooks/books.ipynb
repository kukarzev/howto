```python
import pandas as pd
import logging
logging.getLogger().setLevel(logging.INFO)

# read file
books = pd.read_csv("books.csv")

# select relevant columns
books = books.iloc[:,[0,1,5]]

# rename columns
books.columns = ["title", "author", "score"]

# drop books with 0 and 100 score
books = books.loc[(books.score>0) & (books.score!=100),:]

logging.info("Books considered: %s", len(books))

books
```

```python
import numpy as np

def weighted_choice(objects, weights):
    """ returns randomly an element from the sequence of 'objects', 
        the likelihood of the objects is weighted according 
        to the sequence of 'weights', i.e. percentages."""

    weights = np.array(weights, dtype=np.float64)
    sum_of_weights = weights.sum()
    # standardization:
    np.multiply(weights, 1 / sum_of_weights, weights)
    weights = weights.cumsum()
    x = np.random.random()
    for i in range(len(weights)):
        if x < weights[i]:
            return objects[i]
```

```python
np.random.seed(None)
weighted_choice(books.title,books.score)
```
